## Importe das bibliotecas necessárias

```python
import findspark
findspark.init()
from pyspark.sql import SparkSession

spark = SparkSession.builder \
      .master("local[1]") \
      .appName("SparkByExamples.com") \
      .getOrCreate()
```

## Importe dos dados

```python
clientes = spark.read.load("/content/download/Atividades/Clientes.parquet")

clientes.show(2)
```

```python
clientes.schema
```

```python
vendas = spark.read.load("/content/download/Atividades/Vendas.parquet")

vendas.show(2)
```

### Resolver a primeira questão

```python
atividade1 = clientes.select("Cliente", "Estado", "Status")

atividade1.show()
```

## Resposta da segunda questão

```python
from pyspark.sql import functions as Func

atividade2 = clientes.select("*").where((Func.col("Status") == "Gold") | (Func.col("Status") == "Platinum")). \
                                                                          orderBy(Func.col("ClienteID"))

atividade2.show()
```

### Atividade - 03

```python
from pyspark.sql.functions import sum as SUM

atividade3 = vendas.join(clientes, vendas.ClienteID == clientes.ClienteID).groupBy(clientes.Status). \
                                          agg(SUM("Total")).orderBy(Func.col("sum(Total)").desc())

atividade3.show()
```
